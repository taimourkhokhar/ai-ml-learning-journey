```python
import pandas as pd
df=pd.read_csv("AB_NYC_2019.csv")
df.head()
```

```python
maximum_threshold=df['price'].quantile(0.99)
maximum_threshold
```

```python
df[df['price']>maximum_threshold]
```

```python
min_threshold=df['price'].quantile(0.01)
min_threshold
```

```python
df[df['price']<min_threshold]
```

```python
df[(df['price'] > min_threshold) & (df['price'] < maximum_threshold)]
```
