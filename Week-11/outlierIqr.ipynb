```python
import pandas as pd
df=pd.read_csv("weight-height.csv")
df.head(20)
```

```python
df.describe()
```

```python
q1=df.Height.quantile(0.25)
q1
```

```python
q3=df.Height.quantile(0.75)
q3
```

```python
iqr=q3-q1
upperLimit=q3+1.5*iqr

upperLimit
```

```python
lowerLimit=q1-1.5*iqr
lowerLimit
```

```python
df((df.Height<lowerLimit) | (df.Height>upperLimit) )
```
