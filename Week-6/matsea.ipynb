```python
import subprocess
import sys

def install(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", package])

install('openpyxl')
install('pandas')
install('matplotlib')

# Now try running your code again
import pandas as pd
df = pd.read_excel("linechart.xlsx")
df.head(5)

from matplotlib import pyplot as plt
plt.figure(figsize=(12,7))
plt.plot(df['Quarter'],df['Fridge'],color="orange",label="fridge")
plt.plot(df['Quarter'],df['Dishwasher'],label="dishwasher")
plt.plot(df['Quarter'],df['Washing Machine'],label="washing machine")
plt.title("Product sales")
plt.ylabel("Revenue (min $)")
plt.xlabel("Finacial Quarter")
plt.legend()
# plt.show()
# --- PIE CHART ---
df_sales = df[['Fridge', 'Dishwasher', 'Washing Machine']].sum()
plt.figure(figsize=(8, 8))
plt.pie(df_sales, labels=df_sales.index, autopct="%1.1f%%", explode=(0.1, 0, 0), shadow=True, startangle=140)
plt.title("Total Revenue Contribution")
plt.show()

# --- BAR CHART 1: Totals per Product ---
plt.figure(figsize=(10, 6))
df_sales.plot(kind="bar", color=['orange', 'blue', 'green'])
plt.title("Total Sales by Product")
plt.ylabel("Total Revenue")
plt.xticks(rotation=0) # Keeps product names horizontal
plt.show()

# --- BAR CHART 2: Sales per Quarter (Grouped) ---
# Set Quarter as index so it shows up on the X-axis automatically
df_sales2 = df.set_index("Quarter")
df_sales2.plot(kind="bar", figsize=(12, 7)) 
plt.title("Quarterly Sales Breakdown")
plt.ylabel("Revenue (min $)")
plt.xticks(rotation=45)
plt.legend(title="Products")
plt.show()
#histogram
```
