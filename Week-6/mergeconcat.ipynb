```python
import pandas as pd
india_weather=pd.DataFrame({
  'city':["mumbai","delhi","banglore"],
  'temperature':[32,45,30],
  'humidity':[80,60,78]
})

india_weather

us_weather=pd.DataFrame({
  'city':["new york","chicago","orlando"],
  'temperature':[21,13,35],
  'humidity':[68,65,75]
})

us_weather


df=pd.concat([india_weather,us_weather],keys=["india","usa"])

df.loc['usa']
```

```python
import pandas as pd
df1=pd.DataFrame({
    'city':["new york","chicago","orlando"],
  'temperature':[21,13,35],
})


df2=pd.DataFrame({
  'city':["new york","chicago","orlando"],
  'humidity':[68,65,75]
})

df=pd.merge(df1,df2)
df
```

```python
import pandas as pd
df1=pd.DataFrame({
    'city':["new york","chicago","orlando","baltimore"],
  'temperature':[21,13,35,38],
})


df2=pd.DataFrame({
  'city':["chicago","new york","san diego"],
  'humidity':[68,65,75]
})

df=pd.merge(df1,df2,how="inner")
df
```
