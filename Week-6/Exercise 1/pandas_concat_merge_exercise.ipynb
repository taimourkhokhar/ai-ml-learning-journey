```python
#read the csv file: 'new_movies.csv'
import pandas as pd
df_new_movies=pd.read_csv("movies.csv")

#concat the two dataframes ['df_movies', df_new_movies]
df_movies = pd.concat([df_movies, df_new_movies], ignore_index = True)

#show the last 5 rows
df_movies.tail(5)
```

### <font color="purple">**Question3**</font>

- Merge the dataframe df_movies with the df_languages on 'language_id'.
- Make sure you perform a **inner** join here.
- Store the result again to df_movies

```python
#perform the join operation
import pandas as pd
df_movies=pd.read_csv("movies.csv")
df_languages=pd.read_csv("movies.csv")

#perform the join operation
df_movies = pd.merge(df_movies, df_languages, on = 'language_id', how = 'inner')

#show the top 5 rows
df_movies.head(5)
```
