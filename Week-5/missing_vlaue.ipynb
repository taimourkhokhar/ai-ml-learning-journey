```python
import pandas as pd
df=pd.read_csv("weather_data.csv",parse_dates=["day"])
df.set_index('day',inplace=True)
df.fillna({
  'temperature':df.temperature.mean(),
  'windspeed':df.windspeed.mean(),
  'event':'No Event'
})
```

```python
import pandas as pd
df=pd.read_csv("weather_data.csv",parse_dates=["day"])
df.set_index('day',inplace=True)
df=df.ffill(limit=1)#forward fill
df
```

```python
import pandas as pd
df=pd.read_csv("weather_data.csv",parse_dates=["day"])
df.set_index('day',inplace=True)
df=df.bfill()#backward fill
df
```
