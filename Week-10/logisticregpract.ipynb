```python
from matplotlib import pyplot as plt

left_df = df[df['left'] == 1]

salary_count = left_df['salary'].value_counts()

salary_count.plot(kind='bar')

plt.title("Employees Leaving by Salary")
plt.xlabel("Salary")
plt.ylabel("Number of Employees")
plt.show()
```

```python
left_df = df[df['left'] == 1]
dept_counts = left_df['Department'].value_counts()
import matplotlib.pyplot as plt

dept_counts.plot(kind='bar')
plt.title("Employees Leaving by Department")
plt.xlabel("Department")
plt.ylabel("Number of Employees Left")
plt.show()
```
